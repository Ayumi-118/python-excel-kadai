```python
import openpyxl

# 取引先情報
client = {
    "会社名": "株式会社ABC",
    "住所": "〒101-0022 東京都千代田区神田練塀町300",
    "連絡先": "TEL:03-1234-5678 FAX:03-1234-5678",
    "担当者": "鈴木一郎"
}

# 請求情報

# timeモジュールをインポート
from datetime import datetime

invoice_info = {
    "請求番号": "0001",
    "日付": datetime.today().strftime("%Y/%m/%d")
}

# 取引内容
items = [
    {"商品": "商品A", "数量": 2, "単価": 10000},
    {"商品": "商品B", "数量": 1, "単価": 15000}
]

wb = openpyxl.Workbook()
ws = wb.active

# タイトル
ws["A2"] = "請求書"

# 取引先情報
ws["A4"] = client["会社名"]
ws["A5"] = client["住所"]
ws["A6"] = client["連絡先"]
ws["A7"] = f"担当者名: {client['担当者']} 様"

# 請求番号と日付
ws["E4"] = "No."
ws["F4"] = invoice_info["請求番号"]
ws["E5"] = "日付"
ws["F5"] = invoice_info["日付"]


# 表ヘッダー
ws.append([])
ws.append(["商品名","数量","単価","金額"])

# 商品
total = 0

for item in items:
    amount = item["数量"] * item["単価"]
    ws.append([item["商品"], item["数量"], item["単価"], amount])
    total = total + amount

ws.append(["","","",total])

# 合計
ws.append([])
ws.append(["小計","","",total])
ws.append(["消費税","","",int(total*0.1)])
ws.append(["合計","","",int(total*1.1)])



ws.insert_cols(1, 1)
ws.insert_rows(8, 1)

wb.save("請求書.xlsx")
```
