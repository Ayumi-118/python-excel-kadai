```python
import pandas as pd

file_list = ["2022_年間売上表.xlsx","2023_年間売上表.xlsx"]

df = pd.DataFrame()

for file_name in file_list:
  df_total = pd.read_excel(file_name)

  df_total.columns = ["商品","売上年","金額（千円）"]

  df = pd.concat([df,df_total],ignore_index=True)


```

```python

result= df.groupby(["商品","売上年"],as_index=False)["金額（千円）"].sum()
result.to_excel("売上集計表.xlsx",index=False)
```

```python
import openpyxl
from openpyxl.styles import PatternFill

workbook = openpyxl.load_workbook('売上集計表.xlsx')

sheet = workbook.active

fill = PatternFill(patternType='solid', fgColor='F2F2F2')
for row in sheet['A1:C1']:
    for cell in row:
        cell.fill = fill

workbook.save('売上集計表.xlsx')
```
